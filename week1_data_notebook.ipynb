## Step 1 — Import libraries and load the data

- **pandas** — reads, filters, and summarises tabular data
- **NumPy** — fast math on arrays; pandas uses it under the hood

```python
import pandas as pd
import numpy as np

df = pd.read_csv('movies.csv')
print('Dataset loaded successfully!')
```

## Step 3 — Check for and handle missing values

```python
print('Missing values per column:')
print(df.isna().sum())
```

```python
median_gross = df['gross_million'].median()
df['gross_million'] = df['gross_million'].fillna(median_gross)
print(f'Median gross used to fill: ${median_gross:.1f}M')
print('Missing values after cleaning:')
print(df.isna().sum())
```

---
## Step 4 — Answer 5 questions

### Q1: What is the average IMDb rating across all movies?

```python
avg_rating = df['rating'].mean()
print(f'Average IMDb rating: {avg_rating:.2f} / 10')
print(f'Range: {df["rating"].min()} - {df["rating"].max()}')
```

### Q2: Which genre appears most often?

```python
genre_counts = df['genre'].value_counts()
print(genre_counts.to_string())
print(f'Most common: {genre_counts.index[0]} ({genre_counts.iloc[0]} movies)')
```

### Q3: Which movie made the most money?

```python
top = df.loc[df['gross_million'].idxmax()]
print(f'Movie: {top["title"]} ({top["year"]})')
print(f'Box office: ${top["gross_million"]:.1f} million')
```

### Q4: Top 3 directors by number of movies?

```python
for i, (d, c) in enumerate(df['director'].value_counts().head(3).items(), 1):
    avg = df[df['director']==d]['rating'].mean()
    print(f'{i}. {d}: {c} movies (avg rating {avg:.2f})')
```

### Q5: Do higher-rated movies run longer?

```python
high = df[df['rating'] >= 8.5]
normal = df[df['rating'] < 8.5]
print(f'Rated 8.5+: {high["runtime_min"].mean():.1f} min avg (n={len(high)})')
print(f'Rated <8.5: {normal["runtime_min"].mean():.1f} min avg (n={len(normal)})')
```
